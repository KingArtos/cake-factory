```python
import pandas as pd
```

```python
def percentages(procura):
    total = len(procura)
    return list(map((lambda x: [x, procura.count(x), procura.count(x)/total] ), set(procura)))

def lucro(custo_padrao, feitos, intencao_de_compra, custo_unitario = 5, venda_unitaria = 10):
    return (min(intencao_de_compra, feitos) * venda_unitaria) - custo_padrao - (feitos * custo_unitario)

def apply(df, feitos, custo_unitario = 5, venda_unitaria = 10, custo_fixo = 0):
    df[f"feitos_{feitos}"] = df.apply(lambda row: lucro(custo_fixo, feitos, row.element, custo_unitario, venda_unitaria), axis=1)
    return df

def expected_value(df, producao):
    new_frame = {'feitos': [], 'lucro_esperado': []}
    for feitos in producao:
        new_frame['feitos'].append(feitos)
        new_frame['lucro_esperado'].append(sum(df.apply(lambda row: (row.percent * row[f"feitos_{feitos}"]), axis=1)))
    return new_frame
```

```python
procura = [2,3,0,4,5,2,4,1,0,1,2,3,5,4,5,1,3,0]
producao = [1,2,3,4,5,6,7]
```

onde o primeiro array são as ocorrencias de procura e o segundo array as nossas opções de produção

```python
pd.DataFrame(procura, columns=["procura"])
```

```python
df = pd.DataFrame(percentages(procura), columns=["element", "times", "percent"])
df
```

```python
for feitos in producao:
    df = apply(df, feitos)
df
```

```python
pd.DataFrame(data = expected_value(df, producao))
```
